```python
import os
import cv2 as cv
import numpy as np
import pandas as pd
import glob
```

```python
#main
df = pd.DataFrame(columns=["id", "label", "id"])

count = 0
makedf('0', df)
```

```python
def makedf(class_number, df) :
  os.chdir("/content/gdrive/My Drive/Competition/"+class_number)
  imgs = glob.glob("*")
  for i in range(len(imgs)) :
    image = cv.imread(imgs[i])
    image = cv.resize(image, dsize=(150, 150), interpolation=cv.INTER_LINEAR)
    imlist = list(image.ravel())
    df.loc[i+1, "id"] = imgs[i].split(".")[0]
    df.loc[i+1, "label"] = class_number
    df.loc[i+1, "image"] = imlist
```
